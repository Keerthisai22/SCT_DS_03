```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv('bank.csv')
data['Job']=pd.factorize(data['Job'])[0]
data['Martial']=pd.factorize(data['Martial'])[0]
X=data[['Age','Job','Martial','Balance']]
y=data['purchased']
X_train,X_test=X.iloc[:8],X.iloc[8:]
y_train,y_test=y.iloc[:8],y.iloc[8:]
common=y_train.mode()[0]
y_pred=[common]*len(y_test)
accuracy=np.mean(y_pred==y_test.values)*100
print("Prediction:",y_pred)
print("Actual:",list(y_test.values))
print(f"Accuracy:{accuracy:.2f}%".format(accuracy))
plt.figure(figsize=(3,2))
plt.plot(list(y_test.values),label="Actual",marker='o',color='purple')
plt.plot(y_pred,label="predicted",marker='x',color='red')
plt.title("Actual vs Predicted(purchased)")
plt.show()
         
```
